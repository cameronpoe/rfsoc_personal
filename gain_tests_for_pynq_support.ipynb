```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pynq.overlays.base import BaseOverlay
base = BaseOverlay('base.bit')
base.init_rf_clks()
```

```python
GAIN = 0.6
base.radio.transmitter.channel[0].dac_block.MixerSettings['Freq'] = 1200 # MHz
base.radio.transmitter.channel[0].control.gain = GAIN
base.radio.transmitter.channel[0].control.enable = True

base.radio.transmitter.channel[1].dac_block.MixerSettings['Freq'] = 1200 # MHz
base.radio.transmitter.channel[1].control.gain = GAIN
base.radio.transmitter.channel[1].control.enable = False

NUM_SAMPLES = 32768
times = np.linspace(0, (NUM_SAMPLES - 1)/(4915.2e6/2), NUM_SAMPLES)
data = base.radio.receiver.channel[3].transfer(NUM_SAMPLES)

print(f'Average amplitude over {NUM_SAMPLES} samples: {round(np.average(np.abs(data)),3)} V')
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(times[:256], np.real(data[:256]))
ax.set_title(f'256 Samples with Gain = {GAIN}')
ax.set_ylabel('Voltage (V)')
ax.set_xlabel('Time (s)')
plt.show()
```
